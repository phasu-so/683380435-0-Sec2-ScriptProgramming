```python
num = int(input("Enter a number: ") )
if num < 0:
    text = f"{num} is a negative"
elif num > 0:
    text = f"{num} is a positive"
else:
    text = f"{num} is zero"

if num == 0:
    print(f"{text}.")
elif num % 2 == 0:
    print(f"{text} and an even number.")
else:
    print(f"{text} and an odd number.")
```
