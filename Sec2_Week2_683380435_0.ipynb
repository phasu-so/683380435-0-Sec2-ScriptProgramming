````python
num = int(input("Enter a number: ") )
if num < 0:
    text = f"{num} is a negative"
elif num > 0:
    text = f"{num} is a positive"
else:
    text = f"{num} is zero"

if num == 0:
    print(f"{text}.")
elif num % 2 == 0:
    print(f"{text} and an even number.")
else:
    print(f"{text} and an odd number.")
````

### 1. การกำหนดค่าเริ่มต้น (Initialization)

โปรแกรมเริ่มทำงานโดยเก็บค่าเข้าสู่ตัวแปร 2 ตัว:

* `temp = 25` (ค่าตัวเลขจำนวนเต็ม)
* `is_raining = True` (ค่าตรรกศาสตร์ Boolean)

---

### 2. การตรวจสอบเงื่อนไขทีละขั้นตอน (Step-by-Step Evaluation)

#### **ขั้นตอนที่ 2.1: ตรวจสอบเงื่อนไขแรกที่บรรทัด `if`**

```python
if temp > 20 and not is_raining:

```

* **ซีกซ้าย (`temp > 20`):** แทนค่าได้ `25 > 20` ซึ่งเป็น **`True`**
* **ซีกขวา (`not is_raining`):** แทนค่าได้ `not True` ซึ่งจะกลับประพจน์กลายเป็น **`False`**
* **เชื่อมด้วยตัวดำเนินการ `and`:** เงื่อนไขของ `and` จะเป็นจริงได้*ต้องเป็นจริงทั้งสองฝั่ง* (`True and False`) สรุปแล้วเงื่อนไขในบรรทัด `if` นี้จึงให้ผลลัพธ์รวมเป็น **`False`**
* **ผลลัพธ์:** โปรแกรม**ข้าม**การทำงานของ `print("Perfect day for a picnic!")` ไป

---

#### **ขั้นตอนที่ 2.2: ตรวจสอบเงื่อนไขถัดมาที่บรรทัด `elif`**

เมื่อเงื่อนไขแรกเป็นเท็จ โปรแกรมจะเลื่อนลงมาเช็คเงื่อนไขย่อยถัดไปทันที:

```python
elif temp <= 20 or is_raining:

```

* **ซีกซ้าย (`temp <= 20`):** แทนค่าได้ `25 <= 20` ซึ่งเป็น **`False`**
* **ซีกขวา (`is_raining`):** แทนค่าได้ตรง ๆ คือ **`True`**
* **เชื่อมด้วยตัวดำเนินการ `or`:** เงื่อนไขของ `or` ขอเพียงแค่*มีฝั่งใดฝั่งหนึ่งเป็นจริง* ผลลัพธ์รวมก็จะเป็นจริงทันที (`False or True`) สรุปแล้วเงื่อนไขในบรรทัด `elif` นี้จึงให้ผลลัพธ์รวมเป็น **`True`**
* **ผลลัพธ์:** เงื่อนไขนี้เป็นจริง โปรแกรมจึงเลือก**เข้าไปทำงาน**ในบล็อกนี้และหยุดเช็คเงื่อนไขที่เหลือหลังจากนี้ทันที

---

### 3. ผลลัพธ์ที่โปรแกรมแสดงบนหน้าจอ (Output)

เมื่อโปรแกรมเข้าไปทำงานในบล็อกของ `elif` คำสั่ง `print` จะทำงาน ส่งผลให้หน้าจอแสดงผลลัพธ์ออกมาดังนี้:

```text
Stay indoors or bring an umbrella.

```
